```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas as pd

# Prepare X and y
x = dataframe[['Glucose', 'BMI', 'Insulin', 'SkinThickness', 'DiabetesPedigreeFunction', 'Age']]
y = dataframe['Outcome']

# Standardize the data
sc = StandardScaler()
data = sc.fit_transform(x)

# Convert numpy array to DataFrame for better readability
change = pd.DataFrame(data, columns=x.columns)

# Check the transformed data
print(change)
```

```python
X_train,X_test,y_train,y_test = train_test_split(data,y,test_size=0.2,random_state=55)
```

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score, confusion_matrix
model = Perceptron(max_iter=2000, eta0=0.1, random_state=55)
model.fit(X_train,y_train)
```

```python

# Predict on test data
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Predict on the test set
y_pred = model.predict(X_test)

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)  # Assumes positive class = 1
recall = recall_score(y_test, y_pred)  # Assumes positive class = 1

# Print results
print(f"Accuracy: {accuracy:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
```

```python
from sklearn.metrics import confusion_matrix

# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Print it
print("Confusion Matrix:")
print(cm)
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=["Class 0", "Class 1"], yticklabels=["Class 0", "Class 1"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```
